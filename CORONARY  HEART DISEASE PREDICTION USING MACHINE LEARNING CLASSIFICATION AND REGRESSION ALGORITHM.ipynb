#  Lets  handle  the missing  values

```python
## disply  the missig values  in   each column  
missing_value =df2.isnull().sum()
missing_value
```

```python
##  display the columns that have missing value
missing_value_count=missing_value[missing_value>0]
missing_value_count
```

```python
###  find  the missing  value  percentage  in each column
missing_value_percentage=(missing_value_count*100)/(df2.shape[0])
missing_value_percentage
```

```python
## Find the maximum  missing value percentage
max(missing_value_percentage)
```

```python
# we found that the maximum missing value percentage is 9.15 %.
# so we can use  imputation to fill the missing values
```

```python
from sklearn.impute import SimpleImputer          #It helps fill (or “impute”) missing values in your dataset with some replacement strategy
imputer = SimpleImputer(strategy='most_frequent') #The argument strategy='most_frequent' means:
                                                  #Missing values will be replaced with the most frequent (mode) value in each column.
                                                  #Example:If a column has values [1, 2, 2, NaN, 3], the most frequent value is.
                                                  #The NaN will be replaced with 2.
df3 = pd.DataFrame(imputer.fit_transform(df2))    #does two things: 1. fit(df2): Learns the most frequent value from each column in df2.
                                                  #2. transform(df2): Replaces missing values in df2 with those learned values.
                                                  #The result is a NumPy array.
                                                  #Wrapping it in pd.DataFrame() converts it back to a Pandas DataFrame.
                                                  #The new DataFrame (with no missing values) is stored in df3.
df3.columns = df2.columns  #After imputation, the column names are lost (because fit_transform returns an array).
                           #This line restores the original column names from df2 to df3.
df3.index = df2.index      #That would ensure that row labels (indexes) are the same as in the original DataFrame.
df3.head()
```

# Standardise some features

```python
# Standardise some features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_standardise = ['age','totChol','sysBP','BMI', 'heartRate', \
                       'glucose', 'cigsPerDay']
df3[cols_to_standardise] = scaler.fit_transform(df3[cols_to_standardise])
df3.head()
```

#  Devide the data into  2 parts  one for input  and another for output

```python
X=df3.drop(columns=["TenYearCHD"])
Y=df3.TenYearCHD
```

#  Now  split the data  into training and testing

```python
#   80%  for training and 20% for testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =\
train_test_split(X,Y, test_size=0.2, random_state=40)
```

```python
print(X_train.shape)#(3345, 12)
print(X_test.shape)#(837, 12)
print(Y_train.shape)#(3345,)
print(Y_test.shape)#(837,)
```

# Create  a model   KNeighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
K=KNeighborsClassifier(n_neighbors=7)
# train the model
K.fit(X_train,Y_train)
# test the model
Y_pred_knn=K.predict(X_test)
```

```python
Y_pred_knn
```

```python
Y_test.values
```

```python
##  Find accuracy
from sklearn.metrics import accuracy_score
acc_knn=accuracy_score(Y_test,Y_pred_knn)
acc_knn=round(acc_knn*100,2)
print("Accuracy score in KNN is",acc_knn)
```

```python
##  confusion  matrix
from sklearn.metrics  import  confusion_matrix
confusion_matrix(Y_test,Y_pred_knn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred_knn)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Heart Disease', 'Heart Disease'],
            yticklabels=['No Heart Disease', 'Heart Disease'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - KNN')
plt.show()
```

```python
# classification Report
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred_knn))
```

# Apply  Decsion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
D=DecisionTreeClassifier(min_samples_split=50, random_state=5)
# train the model
D.fit(X_train,Y_train)
# test the model
Y_pred_dt=D.predict(X_test)
##  Find accuracy
from sklearn.metrics import accuracy_score
acc_dt=accuracy_score(Y_test,Y_pred_dt)
acc_dt=round(acc_dt*100,2)
print("Accuracy score in Decisiontree is",acc_dt)
```

```python
##  confusion  matrix
from sklearn.metrics  import  confusion_matrix
confusion_matrix(Y_test,Y_pred_dt)
```

```python
# classification Report
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred_dt))
```
